```python
sentence = input("Enter sentence: ")

# split sentence into words
words = sentence.split()

# convert list to set to get unique words
unique_words = set(words)

# output
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
# Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# initialize
max_val = numbers[0]
min_val = numbers[0]

# loop through list
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

total_days = len(attendance)
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
# Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# Count Character Frequency

text = "python"

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

```python
# Convert List to Tuple

numbers = [10, 20, 30]

result = tuple(numbers)

print("Tuple:", result)
```

```python
# Check if a Key Exists in a Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
