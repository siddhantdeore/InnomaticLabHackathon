```python
def parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1

        if current > peak:
            peak = current

    if current > capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Parking Status:", status)

# Example
parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def delivery_time(distance, traffic, weather):
    time = distance * 5   # base time (5 min per km)

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10

    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

# Example
delivery_time(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    percentage = (booked / total_seats) * 100

    if percentage == 100:
        status = "Housefull"
    elif percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", int(percentage), "%")
    print("Show Status:", status)

# Example
theatre_occupancy(200, [1]*150)
```

```python
def server_load(cpu_list):
    avg = sum(cpu_list) / len(cpu_list)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(avg), "%")
    print("Server Status:", status)

# Example
server_load([45, 60, 70, 85, 90])
```

```python
def classroom_usage(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)

# Example
classroom_usage({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

# Example
event_registration(100, 105)
```
