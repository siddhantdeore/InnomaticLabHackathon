```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def control(self, action):
        self.status = action
        print(f"{self.name} is {self.status}")

# Example
light = SmartLight("Bedroom Light")
light.control("ON")
```

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)

# Example
emp = Employee("Rahul", "EMP102", "IT")
emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

# Example
c = Contact("Anita", "9876543210")
c.display()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

# Example
p = Product("Headphones", 2499)
p.display()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

# Example
m = Movie("Inception", 4.8)
m.display()
```

```python
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)

# Example
d = Delivery("Suman", "Hyderabad")
d.display()
```
